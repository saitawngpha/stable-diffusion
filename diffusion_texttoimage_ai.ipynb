```python
from PIL import Image
from types import SimpleNamespace
from stable_diffusion_tf.stable_diffusion import get_model, text2image
```

```python
args = SimpleNamespace(H=512, W=512, steps=5, scale=7.5, temp=1)
```

```python
text_encoder, diffusion_model, decoder = get_model(args.H, args.W, download_weights=True)
```

```python
prompt = "Shan people dancing on the street at Bangkok"
```

```python
img = text2image(prompt, 
	img_height=args.H, 
	img_width=args.W,  
	text_encoder=text_encoder, 
	diffusion_model=diffusion_model, 
	decoder=decoder,  
	batch_size=1, 
	n_steps=args.steps, 
	unconditional_guidance_scale =args.scale , 
	temperature = 1
)
```

```python
pil_img = Image.fromarray(img[0])
pil_img
```
